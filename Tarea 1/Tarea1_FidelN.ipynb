```python
#Importamos las bibliotecas a utilizar
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

En este programa se evaluaran las funciones:

a) $ a+bx$

b) $a+bx^2$

c) $\sqrt{x}$

d) $\sqrt[3]{x}$

e) $ln(1+2x)$

f) $Sin(x)$

g) $e^{-\frac{x^2}{2}}$

h) $\frac{1}{1+x^2}$

```python
#Establecemos el rango de valores a graficar
x = np.linspace(-20,20,101)
```

```python
#Establecemos el valor de las constantes a y b
A=12.5
B=7.5
```

```python
#Evaluamos las funciones en el rango de valores x
#Los valores se guardarán en un DataFrame cuyas columnas tendran por nombre f_inciso, respetando el orden asignado a las funciones previamente

Datos= pd.DataFrame({
    "x":x,
    "f_a": A + B*x,
    "f_b": A + B*x**2,
    "f_c": np.sqrt(x),
    "f_d": x**(1/3.),
    "f_e": np.log(1 + 2*x),
    "f_f": np.sin(x),
    "f_g": np.exp(-x**2/2),
    "f_h": 1/(1 + x**2)
})

#Visualizamos el dataframe generado
Datos.head(5)

#El mensaje mostrado al correr el código es una advertencia que se esta obteniendo la raíz de un número negativo o calculando el logaritmo de cero.
```

```python
#Guradamos el dataframe como un archivo .csv
Datos.to_csv("Datos.csv",index=False)
```

```python
#Se graficaran las funciones obtenidas
fig, axs = plt.subplots(2,4)
sns.set(rc={'figure.figsize':(15,15)})

axs[0,0].plot(Datos["x"],Datos["f_a"])
axs[0,0].set_title("$A+Bx$", size=20)
axs[0,0].set_ylabel("f(x)")
axs[0,0].set_xlabel("x")

axs[0,1].plot(Datos["x"],Datos["f_b"])
axs[0,1].set_title("$A+Bx^2$", size=20)
axs[0,1].set_xlabel("x")

axs[0,2].plot(Datos["x"],Datos["f_c"])
axs[0,2].set_title("$\sqrt{x}$", size=20)
axs[0,2].set_xlabel("x")

axs[0,3].plot(Datos["x"],Datos["f_d"])
axs[0,3].set_title("$\sqrt[3]{x}$", size=20)
axs[0,3].set_xlabel("x")

axs[1,0].plot(Datos["x"],Datos["f_e"])
axs[1,0].set_title("$ln(1+2x)$", size=20)
axs[1,0].set_ylabel("f(x)")
axs[1,0].set_xlabel("x")

axs[1,1].plot(Datos["x"],Datos["f_f"])
axs[1,1].set_title("$Sin(x)$", size=20)
axs[1,1].set_xlabel("x")

axs[1,2].plot(Datos["x"],Datos["f_g"])
axs[1,2].set_title(r"$e^{-\frac{x^2}{2}}$", size=20)
axs[1,2].set_xlabel("x")

axs[1,3].plot(Datos["x"],Datos["f_h"])
axs[1,3].set_title(r"$\frac{1}{1+x^2}$", size=20)
axs[1,3].set_xlabel("x")

plt.show()
```

<h1>Actividad 2</h1>

En esta actividad se busca elaborar un programa que coloque N partículas (puntos) en una recta de longitud L de forma que la distancia de separación sea uniforme.

```python
#Primero estableceremos la longitud L de nuestra recta
L=10.
#El número de particulas estará dado por la variable N 
N=10.
```

```python
#La posición de cada particula sobre la recta L se obtendra partiendo el intervalo 0:L (siendo 0 el origen) entre las N partículas
def PosP(N,L):
    x = np.linspace(0.,L,N)
    y = np.zeros(int(N))
    return x, y
```

```python
#Usamos la función para generar los puntos
PosX, PosY = PosP(N,L)

#Graficamos los puntos resultantes
plt.figure()
sns.set(rc={'figure.figsize':(5,5)})
plt.scatter(PosX,PosY)
plt.show()
```

<h1>Actividad 3</h1>

Similar a la actividad 2, se realizará un programa que ordene N partículas en una superficie cuadrada de lado L

```python
#Primero estableceremos las medidas de nuestro cuadrado, la medida de su lado estara rado por L
L=20.
#El número de particulas por columna estará dado por la variable N, y el número total de partículas estara dado por N al cuadrado.
N=15
```

```python
#Definimos una función que nos genera la posición de cada partícula.
def Pos2D(L,N):
    x = np.linspace(0.,L,N)
    y = np.linspace(0.,L,N)
    posX = []
    posY = []
    i=j=k=0
    for i in range(int(N)):
        for j in range(int(N)):
            posX.append(x[j])
            posY.append(y[k])
        k=k+1
    Puntos = pd.DataFrame(
    {
        "x" : posX,
        "y" : posY
    })
    
    return Puntos
#La función nos regresa un DataFrame con las posiciones de cada punto
```

```python
#LLamamos a la función para calcular los puntos.
df1 = Pos2D(L,N)
df1.info()
```

Al observar la información del DataFrame con los datos de nuestros puntos tenemos que nos a regresado un total de 225 valores correspondientes a $N^2=15^2=225$.

```python
#Gráficamos los puntos obtenidos
plt.figure()
sns.set(rc={'figure.figsize':(5,5)})
plt.scatter(df1["x"],df1["y"])
plt.show()
```
